# ความยุติธรรมของ AI ในวงการแพทย์

อยากจะชวนทุกคนลองถอดแว่นของแพทย์แล้วมาลองใส่แว่นของนักพัฒนา AI กันดูสักชั่วโมง ในวันนี้เราจะมาคุยกันถึงความรู้พื้นฐานเกี่ยวกับการพัฒนา AI สำหรับการประยุกต์ใช้ทางการแพทย์ แนวคิดเรื่องความยุติธรรม (fairnes) และอคติ (bias) ของการพัฒนา AI ผลกระทบของอคติเหล่านั้น และแนวทางเบื้องต้นในการสร้าง AI ที่มีความยุติธรรม

## ส่วนที่ 1: การสร้างโมเดล AI หมายถึงอะไร

การแก้ปัญหาด้วย AI ไม่มีคำตอบที่สมบูรณ์แบบหรือสูตรสำเร็จ เพราะแต่ละปัญหาต้องใช้แนวคิดและวิธีการที่ต่างกันไปตามรายละเอียด อย่างไรก็ตาม จุดเริ่มต้นคือการทำความเข้าใจปัญหา ข้อมูลที่มีอยู่ และเป้าหมายที่เราต้องการ

ตัวอย่างปัญหาที่พบบ่อยในบริบททางการแพทย์ ได้แก่ การวินิจฉัยผู้ป่วยจากอาการของโรค การคัดกรองโรคจากภาพถ่ายทางการแพทย์ หรือการค้นคว้าวัคซีนใหม่เมื่อเกิดโรคระบาด

แผนภาพด้านล่างแสดงกระบวนการของ machine learning โดยอธิบายขั้นตอนโดยย่อของการสร้างโมเดล AI เพื่อแก้ปัญหาที่กำหนดและบรรลุเป้าหมายที่ตั้งไว้

1. **การเก็บรวบรวมข้อมูล (Data Collection)**

   ขั้นตอนแรกคือการรวบรวมข้อมูลทางการแพทย์ที่เกี่ยวข้องกับปัญหา เช่น ข้อมูลเวชระเบียน ผลการตรวจจากห้องปฏิบัติการ หรือภาพถ่ายทางการแพทย์ นักพัฒนา AI จะต้องเข้าใจปัญหาอย่างลึกซึ้ง โดยเฉพาะในบริบททางการแพทย์ และควรหาข้อมูลหรือทรัพยากรที่เหมาะสมในการนำมาสร้าง AI ให้ได้อย่างมีประสิทธิภาพ

2. **การทำความสะอาดข้อมูล (Data Cleaning)**

   ข้อมูลดิบมักมีข้อผิดพลาด เช่น ค่าที่ขาดหาย เก็บมาไม่ครบ หรือความไม่สอดคล้องกัน เราจึงต้องตรวจสอบและปรับปรุงข้อมูลเหล่านี้อย่างรอบคอบเพื่อให้ข้อมูลมีความถูกต้องและน่าเชื่อถือ

3. **การสร้างฟีเจอร์ (Feature Engineering)**
    เราจะจัดระเบียบข้อมูล และสร้างข้อมูลที่สำคัญที่ไม่ได้มีอยู่ในข้อมูลดิบ เช่น การคำนวณ BMI จากส่วนสูงและน้ำหนัก หรือการติดตามการเปลี่ยนแปลงของค่าตรวจทางห้องปฏิบัติการตามช่วงเวลา ซึ่งทำให้ AI ได้ใช้ข้อมูลที่สำคัญ แทนที่จะให้ AI เห็นแต่ข้อมูลดิบ ในขั้นตอนนี้ เราจะทำให้เกิด *inductive bias* (อคติเชิงอุปนัย) ซึ่งเกิดจากการคาดเดาว่าฟีเจอร์ไหนสำคัญหรือไม่สำคัญ

4. **การฝึกสอนโมเดล (Model Training)**

    โมเดล AI จะเรียนรู้รูปแบบจากข้อมูลและเป้าหมายที่กำหนด หลังจากที่นักพัฒนา AI กำหนดข้อจำกัดหรือกฎเกณฑ์ของข้อมูลเหล่านั้น ตัวอย่างของขั้นตอนนี้เช่น การวิเคราะห์ความสัมพันธ์ระหว่างโปรตีนกับ docking site หรือการจำรูปแบบของภาวะหยุดหายใจขณะหลับกับสัญญาณ SpO2 

5. **การประเมินผล (Evaluation)**

    หลังจากเราสร้างโมเดลขึ้นมาจากข้อมูลแล้ว จะต้องมีการประเมินประสิทธิภาพของโมเดลโดยใช้ตัวชี้วัดต่าง ๆ เช่น sensitivity และ specificity รวมถึงการตรวจสอบอคติเพื่อให้แน่ใจว่าโมเดลมีความยุติธรรมและแม่นยำกับกลุ่มผู้ป่วยที่หลากหลายก่อนนำไปใช้จริง

6. **การนำไปใช้งาน (Deployment)**

    เมื่อโมเดลผ่านการตรวจสอบแล้ว จะนำไปเชื่อมต่อกับระบบของโรงพยาบาล เช่น การแจ้งเตือนผลตรวจที่ผิดปกติ โดยจะต้องมีการฝึกอบรมบุคลากรและจัดให้มีมาตรการติดตามผลอย่างเหมาะสม

7. **การติดตามผล (Monitoring)**

    หลังจากนำไปใช้งานจริง จะต้องมีการติดตามประสิทธิภาพของโมเดลอย่างต่อเนื่อง เช่นเดียวกับแนวทางเวชปฏิบัติที่มีการปรับปรุงอยู่เสมอ โมเดลก็จะได้รับการอัปเดตเมื่อมีข้อมูลใหม่หรือพบแนวทางในการปรับปรุง

กระบวนการนี้คือการสร้าง AI ที่มาสนับสนุนการปฏิบัติงานทางคลินิก โดยยังคงรักษาความปลอดภัยและความน่าเชื่อถือ เพราะความเชี่ยวชาญของแพทย์ยังคงมีความสำคัญในการแปลผลและตัดสินใจขั้นสุดท้าย รวมไปถึงความเข้าใจในปัญหาและการสร้างโมเดล AI 

## ส่วนที่ 2: เริ่มสร้างโมเดล AI ทางการแพทย์ชิ้นแรก

เนื่องจากเราจะพูดถึงความยุติธรรมใน AI กันเป็นเรื่องหลัก เลยอยากทำความเข้าใจกันก่อนว่า “ความยุติธรรม” หมายถึงอะไรในบริบทของ AI 

เพื่อให้เห็นภาพ ลองจินตนาการว่าเรากำลังอยู่ในห้องพิจารณาคดี และผู้พิพากษาในคดีนี้คือระบบ AI หากเราต้องการให้ระบบนี้ “ยุติธรรม” เราย่อมไม่อยากให้ผู้พิพากษาพิจารณาเพศ เชื้อชาติ หรือศาสนาในการตัดสินคดี เพราะเราเชื่อในหลักการที่ว่าผู้ชายและผู้หญิง คนต่างเชื้อชาติ และผู้ที่นับถือศาสนาต่างกัน ควรได้รับการปฏิบัติอย่างเท่าเทียมกันในศาล ดังนั้น เราจึงคาดหวังให้ AI ใช้หลักการนั้นและไม่มีอคติต่อคนกลุ่มใดกลุ่มหนึ่ง

เช่นเดียวกับในสถานการณ์ทางการแพทย์ ลองพิจารณากรณีของโรคมะเร็งปอด หากมีระบบ AI ที่ใช้ช่วยวินิจฉัยโรคมะเร็งปอด เราก็คาดหวังให้ระบบนั้นไม่เลือกปฏิบัติตามเพศ เนื่องจากเราทราบกันดีว่าชายและหญิงมีปัจจัยทางชีวภาพที่เกี่ยวข้องกับโรคนี้ใกล้เคียงกัน เราเรียกโมเดลที่ทำงานโดยปฏิบัติต่อกรณีเหล่านี้อย่างเท่าเทียมว่าเป็นโมเดลที่ยุติธรรม เพราะฉะนั้นความยุติธรรมเลยมีความหมายคล้าย ๆ กับ การไม่ใช้ “ปัจจัยที่ไม่เกี่ยวข้องและไม่เหมาะสม” มาเพื่อเพิ่มความแม่นยำของระบบ

## ส่วนที่ 3: การติดตามผลโมเดล AI ของคุณ

Confusion matrix เป็นเครื่องมือพื้นฐานสำหรับประเมินประสิทธิภาพของ classification model (โมเดลจำแนกประเภท) โดยเป็นตารางที่สรุปจำนวนการทำนายที่ถูกต้องและไม่ถูกต้องของโมเดล แยกตามแต่ละคลาส ตารางนี้แสดงจำนวน true positive (TP), true negative (TN), false positive (FP) และ false negative (FN) ซึ่งช่วยให้เห็นไม่เพียงแต่ความแม่นยำโดยรวม แต่ยังเห็นประเภทของข้อผิดพลาดที่โมเดลทำ การวิเคราะห์นี้ช่วยให้สามารถระบุได้ว่าโมเดลมีอคติต่อคลาสใดหรือมีแนวโน้มทำผิดแบบใดซ้ำ ๆ

<img src="https://www.researchgate.net/publication/377027146/figure/fig5/AS:11431281215285308@1704113587226/The-Matrix-of-TP-FP-FN-TN-Precision-Recall-dan-Accuracy.ppm"
alt="cm-definition" width="400"/>

เราใช้ confusion matrix ในการคำนวนค่าความแม่นยำต่าง ๆ เช่น accuracy, precision, และ recall (sensitivity)

- **Accuracy** วัดความแม่นยำ ซึ่งคือสัดส่วนของการทำนายที่ถูกต้องต่อการทำนายทั้งหมด
- **Precision** วัดความแม่นยำของการทำนายผลที่เป็นบวก นั่นคือสัดส่วนของการทำนายที่ถูกต้องในการทำนายที่เป็นบวกทั้งหมด
- **Recall** วัดความครบถ้วน ซึ่งคือสัดส่วนของการทำนายที่เป็นบวกและถูกต้องต่อจำนวนตัวที่เป็นบวกจริงทั้งหมด

ตัวชี้วัดเหล่านี้ช่วยให้เข้าใจประสิทธิภาพของโมเดลอย่างรอบด้าน โดยเฉพาะในงานทางการแพทย์ที่ข้อผิดพลาดบางประเภทอาจมีผลกระทบอย่างมีนัยสำคัญ

ก่อนที่เราจะเข้าสู่ช่วงถัดไป ลองพิจารณาว่าควรใช้ตัวชี้วัดใดในการประเมินผลสำหรับปัญหาที่เลือกไว้ก่อนหน้านี้ หรือควรมีตัวชี้วัดอื่น ๆ ที่
มีความเหมาะสมมากกว่าตัวชี้วัดเหล่านี้หรือไม่
